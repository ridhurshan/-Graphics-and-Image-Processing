```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# load image
image_path = '20250321_062913.jpg'
img = cv2.imread(image_path)

if img is None:
    print("image not there")
else:
    # hsv convert
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    #greem fine
    lowgreen = np.array([25, 40, 40])
    hightgreen = np.array([90, 255, 255])

    # mask put
    mask = cv2.inRange(hsv, lowgreen, hightgreen)
    kernel = np.ones((5, 5), np.uint8)
    clean_mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
    clean_mask = cv2.morphologyEx(clean_mask, cv2.MORPH_CLOSE, kernel)
    contours, _ = cv2.findContours(clean_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    total_area = 0
    for contour in contours:
        area = cv2.contourArea(contour)
        total_area += area

    # drawline
    output_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    cv2.drawContours(output_img, contours, -1, (255, 0, 0), 20)
    
    plt.imshow(output_img)
    plt.title(f"Leaf Area: {int(total_area)} pixels")
    plt.axis('off')
    plt.show()
    
```

```python
import matplotlib.pyplot as plt

# Leaf area data for 15 days
leaf_areas = [0, 0, 2286, 37840, 62740, 72895, 78043, 93205, 161692, 224916, 266570, 307172, 313701, 334165, ]
days = list(range(1, 15))  # Day numbers from 1 to 14

# Plotting the graph
plt.figure(figsize=(10, 5))
plt.plot(days, leaf_areas, marker='o', linestyle='-', color='green', linewidth=2)

# Add labels and title
plt.title("Leaf Area Growth Over 14 Days", fontsize=14)
plt.xlabel("Day", fontsize=12)
plt.ylabel("Leaf Area (pixels)", fontsize=12)
plt.grid(True)
plt.xticks(days)

# Show plot
plt.tight_layout()
plt.show()
```
